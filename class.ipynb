# クラスとselfと\_\_init\_\_


## はじめに
\_\_init\_\_が難しい理由が2つあります。
- 難しい概念が色々隠れている。
- 自動的に色々やってくれる。

目次
1. オブジェクトって何？
2. クラスって何？
3. インスタンス化って何？
4. \_\_init\_\_って何？
5. selfって何？

## Step 1. オブジェクトって何？
「値」と「処理」をまとめたもの
とら、という猫がいたとする（オブジェクト）
```
# toraは2つの値を持っている
>>> tora.name
'とら'
>>> tora.family
'猫'

# toraは2つの処理を行うことができる
>>> tora.say()
にゃー
>>> tora.growl()
ウー
```

## Step 2. クラスって何？
「共通」の値と処理をまとめたもの   

````python
class Cat:
    family = "猫"

    def say(self):
        print("にゃー")

    def growl(self):
        print("ウー")

tora = Cat()
print(tora.family)
tora.say()
tora.growl()
````

## Step 3. インスタンス化って何？
クラスからオブジェクトを作ること

```
オブジェクト = クラス名()
```

## Step 4. \_\_init\_\_って何？
インスタンス化する時の処理を書く関数

インスタンス化する時に実行する処理を追加するオブジェクトをインスタンス化する時にちょっと改造したいことがあります。

例えば、猫クラスCatをインスタンス化してから名前をつけるのは面倒
```
tora = Cat()
tora.name = 'とら'
```
こうやって1行で書きたい
```
tora = Cat()
```
インスタンス化する時に\_\_init\_\_が自動的に呼び出されるのでこれを使う。

````python
class Cat:
    family = '猫'

    def __init__(self):
        self.name = 'とら'

    def say(self):
        print('にゃー')

    def growl(self):
        print('ウー')


tora = Cat()
tora.name
````

returnは不要。   
関数とは異なり、return文を使わなくても、自動的にselfが返される。

## Step 5. selfって何？

````python
class Cat:
    family = '猫'

    def __init__(生まれたての猫):
        生まれたての猫.name = 'とら'

    def say(self):
        print('にゃー')

    def growl(self):
        print('ウー')


tora = Cat()
tora.name
````

## Step 6. \_\_init\_\_を書き換える
猫の名前を変えてあげられるようにする。   
ここで大事なのは、`__init__(self, name)`メソッドと、`Cat("ドラえもん")`の引数の数が違うこと。

````python
class Cat:
    family = '猫'

    def __init__(self, name):   # <--- __init__の引数は2つ
        self.name = name
    
    def say(self):
        print('にゃー')

    def growl(self):
        print('ウー')


tora = Cat('とら')
print(tora.name)
sachi = Cat('さち')
print(sachi.name)
````

### クラス変数とインスタンス変数
オブジェクトで共有する属性を「クラス変数」、オブジェクトだけで使う属性を「インスタンス変数」と言う。

ここで大事なことはクラス変数にはselfが不要。インスタンス変数にはselfが必要。

````python
class Cat:
    # クラス変数
    # selfが不要
    family = '猫'

    def __init__(self, name):
        # インスタンス変数
        # selfが必要
        self.name = name
    
    def say(self):
        print('にゃー')

    def growl(self):
        print('ウー')


tora = Cat('とら')
print(tora.family)
print(tora.name)
sachi = Cat('さち')
print(sachi.family)
print(sachi.name)
````
