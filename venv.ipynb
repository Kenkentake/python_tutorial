# venv で仮想環境を作る
pip install するディレクトリを分ける（新たに作成する）  
venvを使うと、pip でインストールしたパッケージやモジュールを保存するsite-packagesというディレクトリを新しく作成することが可能。  


sitepackagesの場所は、sys.pathから調べられる。
```
>>> import sys
>>> print(*sys.path, sep="\n")

/usr/local/Cellar/python@3.8/3.8.1/Frameworks/Python.framework/Versions/3.8/lib/python38.zip
/usr/local/Cellar/python@3.8/3.8.1/Frameworks/Python.framework/Versions/3.8/lib/python3.8
/usr/local/Cellar/python@3.8/3.8.1/Frameworks/Python.framework/Versions/3.8/lib/python3.8/lib-dynload
/usr/local/lib/python3.8/site-packages  
```

## 仮想環境の作り方
概要
1. 仮想環境を作成する
2. 仮想環境を有効にする
3. 仮想環境を無効にする
4. 仮想環境を削除する

具体的な手順
1. 作業ディレクトリへ移動
```
cd /Users/user/作業ディレクトリへのパス
```
2. 仮想環境を作成
```
# python3 -m venv 仮想環境名
python3 -m venv venv
# venvディレクトリが作成される
ls -la
>>> vnev
```
3. 仮想環境を有効にする
```
# 仮想環境を有効にする
# コンソールの先頭に(venv)が付く
$ . venv/bin/activate
(venv) $
```
4. 仮想環境を無効にする
```
# 仮想環境を無効にする
# 先頭の(venv)という文字が消える
(venv) $ deactivate
$
```
5. 仮想環境を削除する
```
rm -r venv
```

仮想環境を有効にすると、Pythonのsys.pathとOSの$PATHにvenvへのパスが追加される。  
無効にすると、venvへのパスが削除される。
